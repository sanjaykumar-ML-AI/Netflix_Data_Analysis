### Analyzing 'rating' and 'duration' correlation

To analyze the correlation between `rating` and `duration`, we need to convert the `duration` column into a numerical format. We'll handle 'min' for movies and 'Seasons' for TV shows separately.

```python
# Create a copy to avoid SettingWithCopyWarning
df_processed = df.copy()

# Replace 'Unknown' in duration with NaN before converting to numeric
df_processed.loc[df_processed['type'] == 'Movie', 'duration_numeric'] = df_processed[df_processed['type'] == 'Movie']['duration'].str.replace(' min', '', regex=False).replace('Unknown', np.nan).astype(float)
df_processed.loc[df_processed['type'] == 'TV Show', 'duration_numeric'] = df_processed[df_processed['type'] == 'TV Show']['duration'].str.replace(r' Seasons?', '', regex=True).replace('Unknown', np.nan).astype(float)

# Calculate mean durations for movies and TV shows (excluding NaN)
mean_movie_duration = df_processed[df_processed['type'] == 'Movie']['duration_numeric'].mean()
mean_tvshow_duration = df_processed[df_processed['type'] == 'TV Show']['duration_numeric'].mean()

# Fill NaN values with their respective means
df_processed.loc[(df_processed['type'] == 'Movie') & (df_processed['duration_numeric'].isnull()), 'duration_numeric'] = mean_movie_duration
df_processed.loc[(df_processed['type'] == 'TV Show') & (df_processed['duration_numeric'].isnull()), 'duration_numeric'] = mean_tvshow_duration

print("First 5 rows of movies with numeric duration (after imputation):")
display(df_processed[df_processed['type'] == 'Movie'][['title', 'duration', 'duration_numeric']].head())
print("\nFirst 5 rows of TV Shows with numeric duration (after imputation):")
display(df_processed[df_processed['type'] == 'TV Show'][['title', 'duration', 'duration_numeric']].head())
```

### Average Duration by Rating (with Ordered Ratings)

To better visualize the potential correlation between rating and duration, let's establish a logical order for the `rating` categories and then plot the average `duration_numeric` for each.

```python
# Define a logical order for ratings
rating_order = [
    'G', 'TV-Y', 'TV-Y7', 'TV-G', 'PG', 'TV-PG', 'PG-13',
    'TV-14', 'R', 'TV-MA', 'NC-17', 'Unknown'
]

# Filter for movies, group by rating, and calculate mean duration
mean_movie_duration_by_rating = df_processed[
    df_processed['type'] == 'Movie'
].groupby('rating')['duration_numeric'].mean().reindex(rating_order).dropna()

plt.figure(figsize=(12, 6))
sns.barplot(x=mean_movie_duration_by_rating.index, y=mean_movie_duration_by_rating.values, palette='coolwarm')
plt.title('Average Movie Duration (minutes) by Rating (Ordered)')
plt.xlabel('Rating')
plt.ylabel('Average Duration (minutes)')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Filter for TV shows, group by rating, and calculate mean duration
mean_tvshow_seasons_by_rating = df_processed[
    df_processed['type'] == 'TV Show'
].groupby('rating')['duration_numeric'].mean().reindex(rating_order).dropna()

plt.figure(figsize=(12, 6))
sns.barplot(x=mean_tvshow_seasons_by_rating.index, y=mean_tvshow_seasons_by_rating.values, palette='plasma')
plt.title('Average TV Show Seasons by Rating (Ordered)')
plt.xlabel('Rating')
plt.ylabel('Average Number of Seasons')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

### Analyzing Director and Cast Relationships

To find the relationships between directors and cast members, we need to handle the comma-separated strings in both columns. We will create a new DataFrame where each row represents a unique director-cast pair for a given title, excluding 'Unknown' entries which were used for imputation.

```python
import itertools

# Filter out 'Unknown' directors and cast members for this analysis
df_relations = df_processed[(df_processed['director'] != 'Unknown') & (df_processed['cast'] != 'Unknown')].copy()

# Function to split and clean string entries (e.g., remove leading/trailing spaces)
def split_and_clean(text):
    if isinstance(text, str):
        return [item.strip() for item in text.split(',')]
    return []

# Apply the function to director and cast columns
df_relations['director_list'] = df_relations['director'].apply(split_and_clean)
df_relations['cast_list'] = df_relations['cast'].apply(split_and_clean)

# Create a new DataFrame with one row per director-cast pair
# This involves exploding both lists, which requires a bit more logic than a single explode

# First, explode directors
df_directors_exploded = df_relations.explode('director_list')

# Now, prepare to explode cast members for each director-title combination
# We'll re-merge to ensure the correct cast_list is associated with each director_list entry
df_director_cast = df_directors_exploded.groupby(['show_id', 'director_list'])['cast_list'].apply(lambda x: list(itertools.chain.from_iterable(x))).reset_index()
df_director_cast = df_director_cast.explode('cast_list')

# Rename columns for clarity
df_director_cast.rename(columns={'director_list': 'director_name', 'cast_list': 'cast_member'}, inplace=True)

# Drop duplicates in case a director and cast member worked on the same title and were listed multiple times
df_director_cast.drop_duplicates(inplace=True)

print("First 5 rows of director-cast relationships:")
display(df_director_cast.head())
```

#### Top Directors by Number of Productions

Let's see which directors have the most entries in the filtered dataset (where cast is also known).

```python
top_directors = df_relations['director_list'].explode().value_counts().head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x=top_directors.index, y=top_directors.values, palette='Greens_d')
plt.title('Top 10 Directors by Number of Productions (with known cast)')
plt.xlabel('Director')
plt.ylabel('Number of Productions')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

#### Top Cast Members by Number of Productions

Similarly, let's identify the most prolific cast members in the filtered dataset.

```python
top_cast = df_relations['cast_list'].explode().value_counts().head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x=top_cast.index, y=top_cast.values, palette='Blues_d')
plt.title('Top 10 Cast Members by Number of Productions (with known director)')
plt.xlabel('Cast Member')
plt.ylabel('Number of Productions')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

#### Frequent Director-Cast Collaborations

Now, let's look at which directors and cast members frequently work together. We'll find the top director-cast pairs that appear in multiple productions.

```python
collaboration_counts = df_director_cast.groupby(['director_name', 'cast_member']).size().reset_index(name='collaboration_count')

# Filter for collaborations that happened more than once
frequent_collaborations = collaboration_counts[collaboration_counts['collaboration_count'] > 1].sort_values(by='collaboration_count', ascending=False).head(10)

print("Top 10 Frequent Director-Cast Collaborations:")
display(frequent_collaborations)
```

### Relationship Between Director and Content Type

Let's explore which directors are more inclined towards directing movies versus TV shows.

```python
# Filter out 'Unknown' directors
df_filtered_directors = df_processed[df_processed['director'] != 'Unknown'].copy()

# Split directors string into individual names
df_filtered_directors['director_name'] = df_filtered_directors['director'].apply(lambda x: [d.strip() for d in x.split(',')])

# Explode the director_name column to have one director per row
df_exploded_directors = df_filtered_directors.explode('director_name')

# Group by director and type, then count productions
director_type_counts = df_exploded_directors.groupby(['director_name', 'type']).size().unstack(fill_value=0)

# Calculate total productions for each director
director_type_counts['Total'] = director_type_counts['Movie'] + director_type_counts['TV Show']

# Sort by total productions and display the top 10
top_directors_by_total = director_type_counts.sort_values(by='Total', ascending=False).head(10)

print("Top 10 Directors by Total Productions (Movies and TV Shows):")
display(top_directors_by_total)
```

#### Top Directors by Number of Movies

Let's visualize the directors who have directed the most movies.

```python
top_movie_directors = director_type_counts.sort_values(by='Movie', ascending=False).head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x=top_movie_directors.index, y=top_movie_directors['Movie'], palette='Reds_d')
plt.title('Top 10 Directors by Number of Movies')
plt.xlabel('Director')
plt.ylabel('Number of Movies')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

#### Top Directors by Number of TV Shows

Now, let's visualize the directors who have directed the most TV shows.

```python
top_tvshow_directors = director_type_counts.sort_values(by='TV Show', ascending=False).head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x=top_tvshow_directors.index, y=top_tvshow_directors['TV Show'], palette='Blues_d')
plt.title('Top 10 Directors by Number of TV Shows')
plt.xlabel('Director')
plt.ylabel('Number of TV Shows')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

### Relationship Between Director and Duration

Let's analyze the average duration of content (movies in minutes, TV shows in seasons) for each director.

```python
# Group by director name and type, then calculate the mean duration
director_avg_duration = df_exploded_directors.groupby(['director_name', 'type'])['duration_numeric'].mean().unstack()

print("First 10 directors and their average movie/TV show duration:")
display(director_avg_duration.head(10))
```

## Summary

Here is a final summary of the key insights and analysis findings from the Netflix dataset, presented in 12 points:

1) Missing Value Handling: Missing values in 'director', 'cast', 'country', 'rating', 'duration', and 'date_added' columns were successfully imputed, primarily with 'Unknown' for categorical features and mean values for numerical durations.

2) Duration Standardization: The 'duration' column was cleaned and converted into a numerical 'duration_numeric' (minutes for movies, seasons for TV shows), and remaining missing durations were imputed with the respective mean values for movies and TV shows.

3) Content Type Distribution: The dataset shows a predominant number of movies compared to TV shows, indicating a higher volume of film content on Netflix.

4) Release Year Trends: The release year distribution highlights a significant increase in content production, particularly in recent years (post-2010), with a peak around 2017-2019.

5) Movie Duration by Rating: The average movie duration varies across different ratings, with some ratings like 'NC-17' and 'PG-13' having generally longer average film lengths, and 'TV-Y' and 'G' having shorter durations.

6) TV Show Seasons by Rating: TV shows generally have fewer seasons on average, with little significant variation across different content ratings, most falling within 1-2 seasons.

7) Top Content-Producing Countries: The United States, India, and the United Kingdom are the leading countries in content production on Netflix, both in terms of overall volume and proportion of content.

8) Content Categories: The word cloud of 'listed_in' categories revealed prominent genres such as 'Dramas', 'International Movies', 'Comedies', and 'Documentaries', reflecting popular content types on the platform.

9) Top Prolific Directors: Directors like Rajiv Chilaka, Jan Suter, and Raúl Campos emerge as highly prolific, having directed a substantial number of productions (movies or TV shows) in the dataset.

10) Top Prolific Cast Members: Anupam Kher, Shah Rukh Khan, and Naseeruddin Shah are among the most frequently appearing cast members, indicating their widespread involvement in Netflix productions.

11) Frequent Director-Cast Collaborations: Specific director-cast pairs, such as Rajiv Chilaka and Rajesh Kava, show high collaboration counts, suggesting established working relationships.

12) Director Specialization in Duration: Some directors exhibit specialization in content duration; for instance, Houssam El-Din Mustafa and Samir Al Asfory tend to direct exceptionally long movies, while Phil Sgriccia and Andy Devonshire are associated with TV shows having a higher average number of seasons.
